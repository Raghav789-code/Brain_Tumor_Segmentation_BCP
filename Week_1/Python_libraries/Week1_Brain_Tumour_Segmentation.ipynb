# Assignment 1
'''
Create a numpy array of length 1000 which resembles a normal distribution with mean 0 and standard deviation 1.
Create 2 subplots both histograms with 20 bins side by side, one which has a beta (a=2, b=5) distribution, the other which has the normal distribution.
'''

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
normal_data = np.random.normal(loc=0, scale=1, size=1000)
beta_data = np.random.beta(a=2, b=5, size=1000)

# Create side-by-side histograms
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(beta_data, bins=20)
plt.title("Beta Distribution (a=2, b=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(normal_data, bins=20)
plt.title("Normal Distribution (=0, =1)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

# Assignment 2
'''
Generate a NumPy array of 1000 random values drawn from a normal distribution with mean 0 and std. dev 1
Plot a histogram of the data with 20 bins.
Add vertical lines to indicate the mean of the data, the 25th percentile and the 75th percentile.
Annotate the plot with labels for the mean, 25th percentile, and 75th percentile.
Customize the visualization (light pink colour scheme, titles, labels, legend).
'''

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data = np.random.normal(loc=0, scale=1, size=1000)

# Compute statistics
mean_val = np.mean(data)
p25 = np.percentile(data, 25)
p75 = np.percentile(data, 75)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=20, color="lightpink", edgecolor="black", alpha=0.7)

# Add vertical lines
plt.axvline(mean_val, color="red", linestyle="--", label=f"Mean = {mean_val:.2f}")
plt.axvline(p25, color="purple", linestyle="--", label=f"25th Percentile = {p25:.2f}")
plt.axvline(p75, color="darkmagenta", linestyle="--", label=f"75th Percentile = {p75:.2f}")

# Labels and title
plt.title("Histogram of Normal Distribution with Key Statistics")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.show()

```

# Assignment 1
'''
Create an array of 18 random elements of shape (3, 6) and add the array [0, 0, 2, 4, 5, 3] to each row.
Then reshape it to a (9, 2) array and take its transpose.
Find the locations where all elements are greater than the mean of the array and print it
'''



```python
import numpy as np

# Step 1: Create random array
arr = np.random.rand(3, 6)

# Step 2: Add array to each row
add_arr = np.array([0, 0, 2, 4, 5, 3])
arr_added = arr + add_arr

# Step 3: Reshape to (9, 2)
reshaped = arr_added.reshape(9, 2)

# Step 4: Transpose
transposed = reshaped.T

# Step 5: Find mean
mean_val = transposed.mean()

# Step 6: Find locations where elements > mean
locations = np.where(transposed > mean_val)

print("Original array (3x6):\n", arr)
print("\nAfter adding [0, 0, 2, 4, 5, 3]:\n", arr_added)
print("\nReshaped to (9, 2):\n", reshaped)
print("\nTransposed array:\n", transposed)
print("\nMean of transposed array:", mean_val)
print("\nLocations where elements > mean:", locations)
```

# Assignment 2
'''
Create an array of length 20 that models a poisson distribution and center it around the mean and normalise it by dividing it by the standard deviation
(Centering around some value "a" is subtracting "a" from each element and normalising is to divide it by some factor)
'''

```python
import numpy as np

# Step 1: Generate Poisson distribution
poisson_data = np.random.poisson(lam=5, size=20)

# Step 2: Compute mean and standard deviation
mean_val = poisson_data.mean()
std_val = poisson_data.std()

# Step 3: Center and normalize
normalized_data = (poisson_data - mean_val) / std_val

print("Original Poisson data:\n", poisson_data)
print("\nMean:", mean_val)
print("Standard Deviation:", std_val)
print("\nCentered and normalized data:\n", normalized_data)
```

```python
 # Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the video game sales dataset
# Make sure videogamesales.csv is uploaded to your Colab environment
df = pd.read_csv("videogamesales.csv")

# Create a new column 'global_sales' by summing all regional sales
df["global_sales"] = (
    df["NA_Sales"] +
    df["EU_Sales"] +
    df["JP_Sales"] +
    df["Other_Sales"]
)

# Sort the DataFrame based on global sales in descending order
df_sorted = df.sort_values(by="global_sales", ascending=False)

# Display the top 10 games by global sales
print("Top 10 Games by Global Sales")
display(df_sorted.head(10))

# Group the data by genre and calculate total global sales for each genre
genre_sales = df.groupby("Genre")["global_sales"].sum()

# Plot total number of copies sold of each genre globally
plt.figure(figsize=(10, 6))
genre_sales.plot(kind="bar")
plt.title("Total Global Sales by Genre")
plt.xlabel("Genre")
plt.ylabel("Copies Sold (Millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Filter games that contain 'Grand Theft Auto' in their name
gta_df = df[df["Name"].str.contains("Grand Theft Auto", case=False, na=False)]

# Create a new column for combined Europe and Japan sales
gta_df["EU_JP_Sales"] = gta_df["EU_Sales"] + gta_df["JP_Sales"]

# Select only the required columns for display
gta_filtered = gta_df[["Name", "Platform", "Year", "EU_JP_Sales"]]

# Display the filtered Grand Theft Auto DataFrame
print("Grand Theft Auto Games Data")
display(gta_filtered)

# Calculate total sales of all GTA games by region
gta_sales_region = {
    "North America": gta_df["NA_Sales"].sum(),
    "Europe": gta_df["EU_Sales"].sum(),
    "Japan": gta_df["JP_Sales"].sum(),
    "Other": gta_df["Other_Sales"].sum()
}

# Plot a pie chart showing regional sales distribution of GTA games
plt.figure(figsize=(7, 7))
plt.pie(
    gta_sales_region.values(),
    labels=gta_sales_region.keys(),
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Regional Sales Distribution of Grand Theft Auto Games")
plt.show()
```
